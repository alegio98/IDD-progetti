```python
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('ELETTRONICA','Ingegneria Elettronica') #sostituisco ELETTRONICA con Ingegneria Elettronica
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('CIVILE','Ingegneria Civile') 
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('EDILE','Ingegneria Edile') 
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('MECCANICA','Ingegneria Meccanica') 
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('INFORMATICA','Ingegneria Informatica') 
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('ELETTRICA','Ingegneria Elettrica') 
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('DELLE TELECOMUNICAZIONI','Ingegneria Delle Telecomunicazioni') 
professionisti['Specializzazione'] = professionisti['Specializzazione'].str.replace('NUCLEARE','Ingegneria Energetica e Nucleare') 
```

```python
ax = (professionisti['Specializzazione'].value_counts()*100.0 /len(professionisti))\
.plot.pie(autopct='%.1f%%',figsize =(16,16), fontsize = 10 )                                                                           
#ax.yaxis.set_major_formatter(mtick.PercentFormatter())
#ax.set_ylabel('Settori',fontsize = 12)
ax.set_title('% of Specializzazioni', fontsize = 12)
```
